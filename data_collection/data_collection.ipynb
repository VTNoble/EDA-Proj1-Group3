```python
# Dependencies
import requests
import json
import pandas as pd

# Import API keys
from api_keys import api_key
from api_keys import secret
```

```python
# necessary to pass for a cURL command in jupyter notebook
data = f"grant_type=client_credentials&client_id={api_key}&client_secret={secret}"

# necessary to pass for a cURL command in jupyter notebook 
headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
}

# base url for token request
token_url = f'https://api.petfinder.com/v2/oauth2/token'
```

```python
# token request
token = requests.post(token_url, headers=headers, data=data).json()
```

```python
# create headers variable for data request
headers = {"Authorization": f"{token['token_type']} {token['access_token']}"}
```

```python
# create empty lists of data to collect
pet_id = []
name = []
status = []
publish_date = []
org_id = []
city = []
state = []
breed = []
age = []
gender = []
size = []
coat = []
spayed_neutered = []
house_trained = []
special_needs = []
shots_current = []

# loop to gather data from API with separate API calls for individual random dogs in Georgia
for i in range(1000):
    
    # construct data request with access token
    url = f'https://api.petfinder.com/v2/animals?type=dog&location=georgia&sort=random&limit=1'
    
    # get response and jsonify
    response = requests.get(url, headers=headers).json()

    # if pet_id from response is not in pet_id list, append the data
    if response['animals'][0]['id'] not in pet_id:
        try:
            pet_id.append(response['animals'][0]['id'])
            name.append(response['animals'][0]['name'])
            status.append(response['animals'][0]['status'])
            publish_date.append(response['animals'][0]['published_at'])
            org_id.append(response['animals'][0]['organization_id'])
            breed.append(response['animals'][0]['breeds']['primary'])
            age.append(response['animals'][0]['age'])
            gender.append(response['animals'][0]['gender'])
            size.append(response['animals'][0]['size'])
            coat.append(response['animals'][0]['coat'])
            spayed_neutered.append(response['animals'][0]['attributes']['spayed_neutered'])
            house_trained.append(response['animals'][0]['attributes']['house_trained'])
            special_needs.append(response['animals'][0]['attributes']['special_needs'])
            shots_current.append(response['animals'][0]['attributes']['shots_current'])
            city.append(response['animals'][0]['contact']['address']['city'])
            state.append(response['animals'][0]['contact']['address']['state'])
            print(f'Dog was found! Appending data...')
        except:
            print(f'Dog was not found!')

    # if pet_id from response is already in pet_id list, skip it
    else:
        pass
```

```python
# verify sample size by checking length of pet_id
len(pet_id)
```

```python
# create dictionary to pass into DataFrame
dog_dict = {
    'Pet ID': pet_id,
    'Name': name,
    'Status': status,
    'Publish Date': publish_date,
    'Organization ID': org_id,
    'City': city,
    'State': state,
    'Breed': breed,
    'Age': age,
    'Gender': gender,
    'Size': size,
    'Coat': coat,
    'Spayed or Neutered': spayed_neutered,
    'House Trained': house_trained,
    'Special Needs': special_needs,
    'Shots Current': shots_current,
}

# create DataFrame from dictionary
georgia_dogs = pd.DataFrame(dog_dict)
georgia_dogs.head()
```
